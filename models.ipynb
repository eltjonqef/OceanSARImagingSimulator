```python
import os
output_folder="figures"
if not os.path.exists(output_folder):
        os.makedirs(output_folder)
```

```python
import numpy as np
from omnidirectional_spectrum import omnidirectional_spectrum, spectrum_model

wavenumbers=np.logspace(-3,5,500)
windspeeds=[4,8,12,16]
```

```python
PM=[omnidirectional_spectrum(spectrum=spectrum_model.Pierson_Moskowitz,k=wavenumbers,v=windspeed) for windspeed in windspeeds]
for model in PM:
    model.plot(f'{output_folder}/pierson_moskowitz.png',[0.004,16000])
```

```python
JONSWAP=[omnidirectional_spectrum(spectrum=spectrum_model.JONSWAP,k=wavenumbers,v=windspeed,F=25000) for windspeed in windspeeds]
for model in JONSWAP:
    model.plot(f'{output_folder}/JONSWAP.png',[0.03,17000])
```

```python
Elfouhaily_spectrum=[omnidirectional_spectrum(spectrum=spectrum_model.Elfouhaily,k=wavenumbers,v=windspeed,F=25000) for windspeed in windspeeds]
for model in Elfouhaily_spectrum:
    model.plot(f'{output_folder}/Elfouhaily_Spectrum.png',[0.004,2500])
```

```python
import numpy as np
from spreading_function import spreading_function, spreading_model
theta=np.linspace(0, 2*np.pi, 200)
```

```python
n=[2,4,6,8]
simple_cosine=[spreading_function(function=spreading_model.Simple_Cosine, theta=theta, wind_direction=np.pi/3, n=i) for i in n]
for model in simple_cosine:
    model.plot(f'{output_folder}/simple_cosine.png')
```

```python
S=[1,3,5,10]
LH=[spreading_function(function=spreading_model.Longuet_Higgins,theta=theta, wind_direction=np.pi/3,S=s) for s in S]
for model in LH:
    model.plot(f'{output_folder}/Longuet_Higgins.png')
```

```python
K=[202.68, 110.23, 26.73, 0.1]
Elfouhaily_spreading=[spreading_function(function=spreading_model.Elfouhaily, theta=theta, wind_direction=np.pi/3,k=k, v=8) for k in K]
for model in Elfouhaily_spreading:
    model.plot(f'{output_folder}/Elfouhaily_Spreading.png')
```
