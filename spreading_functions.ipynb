```python
import numpy as np
import spreading_functions
import plots
wavenumbers=np.linspace(0.001,10000,10000000)
theta=2*np.pi*np.linspace(0,2,200)
windspeeds=[4]#,8,12,16]
```

Simple Cosine Function

```python
n=[2,4,6,8]
Simple_Cosine_Spreads=[spreading_functions.Simple_Cosine(theta, i) for i in n]
```

```python
plots.Spread(theta, n, Simple_Cosine_Spreads, "Simple Cosine","n","SimpleCosine")
```

Longuet-Higgins

```python
S=[8, 10, 20]
LH_Spreads=[spreading_functions.Longuet_Higgins(theta, i) for i in S]
```

```python
plots.Spread(theta, S, LH_Spreads, "Longuet-Higgins","S", "Longuet_Higgins")
```

Elfouhaily

```python
Elfouhaily_Spread=[spreading_functions.Elfouhaily(theta, 0,1, windspeed) for windspeed in windspeeds]
```
